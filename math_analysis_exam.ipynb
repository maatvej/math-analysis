$ y = e^{x \ln(3x + 1)} $
Применение правила производной сложной функции

Для функции вида $ y = e^{u} $ производная будет $ y' = e^{u} u' $, где $ u = x \ln(3x + 1) $
Находим $ u' $

Для $ u = x \ln(3x + 1) $ используем правило произведения:

$
u' = \frac{d}{dx}(x) \cdot \ln(3x + 1) + x \cdot \frac{d}{dx}\left(\ln(3x + 1)\right)
$

Вычислим $ \frac{d}{dx}\left(\ln(3x+1)\right) $

$
\frac{d}{dx}\left(\ln(3x + 1)\right) = \frac{3}{3x + 1} \quad 
$

Таким образом:

$
u' = \ln(3x + 1) + x \cdot \frac{3}{3x + 1}
$
$
= \ln(3x + 1) + \frac{3x}{3x + 1}
$

Записываем значение производной
$
y' = e^{x \ln(3x + 1)} \left( \ln(3x + 1) + \frac{3x}{3x + 1} \right)
$

### ЗАдание 2
Для нахождения безусловных экстремумов функции $ u = 3x^3 + y^2 + z^2 + 6xy - 2z + 1 $ мы будем использовать метод нахождения критических точек и вторичный тест для экстремумов.

### Вычисление частных производных

Сначала найдем частные производные функции $ u $ по переменным $ x $, $ y $, и $ z $:
- **Частная производная по $ x $**:

   $
   \frac{\partial u}{\partial x} = 9x^2 + 6y
   $
- **Частная производная по $ y $**:

   $
   \frac{\partial u}{\partial y} = 2y + 6x
   $
- **Частная производная по $ z $**:

   $
   \frac{\partial u}{\partial z} = 2z - 2
   $

### Находим критические точки

Теперь приравняем каждую из частных производных к нулю и решим систему уравнений.

$
9x^2 + 6y = 0 \quad \text{(1)}
$

$
2y + 6x = 0 \quad \text{(2)}
$

$
2z - 2 = 0 \quad \Rightarrow \quad z = 1 \quad \text{(3)}
$

### Решаем систему уравнений

Из уравнения (3) мы получили значение $ z = 1 $.

Теперь решим систему (1) и (2):

Из уравнения (2):
$
2y + 6x = 0 \Rightarrow y = -3x \quad \text{(4)}
$

Подставим (4) в (1):

$
9x^2 + 6(-3x) = 0
$
$
9x^2 - 18x = 0
$
$
9x(x - 2) = 0
$

Таким образом, у нас есть два случая:
- $ x = 0 $

- $ x = 2 $

Теперь находим $ y $ для каждого значения $ x $:
- Если $ x = 0 $:

  $
  y = -3(0) = 0
  $
- Если $ x = 2 $:

  $
  y = -3(2) = -6
  $

Теперь мы имеем следующие критические точки:
- $ (0, 0, 1) $

- $ (2, -6, 1) $

### Проверка второго производного теста

Теперь нам нужно вычислить вторые производные $ u $ и построить матрицу Гессе.

### Вторые производные:

$
\frac{\partial^2 u}{\partial x^2} = 18x
$

$
\frac{\partial^2 u}{\partial y^2} = 2
$

$
\frac{\partial^2 u}{\partial z^2} = 2
$

$
\frac{\partial^2 u}{\partial x \partial y} = 6
$

$
\frac{\partial^2 u}{\partial y \partial z} = 0
$

$
\frac{\partial^2 u}{\partial z \partial x} = 0
$

Теперь составим матрицу Гессе $ H $:

$$
H = \begin{bmatrix}
\frac{\partial^2 u}{\partial x^2} & \frac{\partial^2 u}{\partial x \partial y} & \frac{\partial^2 u}{\partial x \partial z} \\
\frac{\partial^2 u}{\partial y \partial x} & \frac{\partial^2 u}{\partial y^2} & \frac{\partial^2 u}{\partial y \partial z} \\
\frac{\partial^2 u}{\partial z \partial x} & \frac{\partial^2 u}{\partial z \partial y} & \frac{\partial^2 u}{\partial z^2}
\end{bmatrix}
=
\begin{bmatrix}
18x & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{bmatrix} 
$$

### Проверка в критической точке $ (0, 0, 1) $

Подставим $ x = 0 $:

$
H(0, 0, 1) = \begin{bmatrix}
0 & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{bmatrix}
$

Теперь вычислим определитель матрицы $ H $.
- Определитель мини-матрицы $ 2 \times 2 $:

$
M_2 = \begin{bmatrix}
0 & 6 \\
6 & 2
\end{bmatrix} \Rightarrow \det(M_2) = (0)(2) - (6)(6) = -36 Следовательно, $ (0, 0, 1) $ является седловой точкой, и это не экстремум.

### Проверка в критической точке $ (2, -6, 1) $

Подставим $ x = 2 $:

$
H(2, -6, 1) = \begin{bmatrix}
36 & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{bmatrix}
$

Теперь вычислим определитель матрицы $ H $.
- Определитель матрицы:

$
\det(H) = 36(2 \cdot 2) - 6(6 \cdot 0) + 0 = 144
$
Положительное значение первого определителя и положительные значения остальных ведущих миноров указывают на то, что точка $ (2, -6, 1) $ является локальным минимумом.

### Заключение

Таким образом, функция $ u = 3x^3 + y^2 + z^2 + 6xy - 2z + 1 $ имеет локальный минимум в точке $ (2, -6, 1) $.

### Задание 3
Для нахождения условного экстремума функции $ z = e^{-x - y} $ при условии $ g(x, y) = \frac{x^2}{4} + y^2 - 5 = 0 $, мы будем использовать метод множителей Лагранжа. Этот метод заключается в составлении функции Лагранжа и нахождении её экстремумов с учетом заданного условия.

### Составляем функцию Лагранжа

Функция Лагранжа имеет вид:
$
\mathcal{L}(x, y, \lambda) = e^{-x - y} + \lambda \left(\frac{x^2}{4} + y^2 - 5\right)
$

### Находим частные производные и приравниваем их к нулю

Теперь найдем частные производные функции Лагранжа:
- По $ x $:

   $
   \frac{\partial \mathcal{L}}{\partial x} = -e^{-x - y} + \lambda \frac{x}{2} = 0
   $
- По $ y $:

   $
   \frac{\partial \mathcal{L}}{\partial y} = -e^{-x - y} + 2\lambda y = 0
   $
- По $ \lambda $:

   $
   \frac{\partial \mathcal{L}}{\partial \lambda} = \frac{x^2}{4} + y^2 - 5 = 0
   $

### Решаем систему уравнений

Теперь у нас есть система из трех уравнений:
- $-e^{-x - y} + \lambda \frac{x}{2} = 0 $

- $-e^{-x - y} + 2\lambda y = 0 $

- $\frac{x^2}{4} + y^2 - 5 = 0 $

Из первых двух уравнений выразим $ \lambda $:
- Из первого:

   $
   \lambda = \frac{2e^{-x - y}}{x} \quad (x \neq 0)
   $
- Из второго:

   $
   \lambda = \frac{e^{-x - y}}{2y} \quad (y \neq 0)
   $

Теперь приравняем два выражения для $ \lambda $:
$
\frac{2e^{-x - y}}{x} = \frac{e^{-x - y}}{2y}
$

### Шаг 4: Упрощаем и решаем уравнение

При $ e^{-x - y} \neq 0 $ можно сократить:
$
\frac{2}{x} = \frac{1}{2y}
$

Преобразуем это уравнение:
$
4y = x
$
Отсюда:
$
x = 4y
$

### Подставляем $ x $ в условие

Теперь подставим $ x = 4y $ в условие $ \frac{x^2}{4} + y^2 = 5 $:
$
\frac{(4y)^2}{4} + y^2 = 5
$
$
\frac{16y^2}{4} + y^2 = 5
$
$
4y^2 + y^2 = 5
$
$
5y^2 = 5 \quad \Rightarrow \quad y^2 = 1 \quad \Rightarrow \quad y = 1 \quad \text{или} \quad y = -1
$

### Находим соответствующие значения $ x $

Теперь найдем $ x $ для каждого $ y $:
- Если $ y = 1 $:

   $
   x = 4 \cdot 1 = 4
   $
- Если $ y = -1 $:

   $
   x = 4 \cdot (-1) = -4
   $

### Находим значения функции $ z $

Теперь мы имеем два кандидата для оптимизации:
- $ (4, 1) $

- $ (-4, -1) $

Вычислим значение функции $ z = e^{-x - y} $ в обеих точках:
- Для точки $ (4, 1) $:

   $
   z(4, 1) = e^{-4 - 1} = e^{-5}
   $
- Для точки $ (-4, -1) $:

   $
   z(-4, -1) = e^{-(-4) - (-1)} = e^{4 - 1} = e^{3}
   $

### Сравнение значений

- Значение $ z(4, 1) = e^{-5} $ — это минимальное значение, так как оно меньше нуля.

- Значение $ z(-4, -1) = e^{3} $ — максимальное значение.

### Заключение

Условный экстремум функции $ z = e^{-x - y} $ при условии $ \frac{x^2}{4} + y^2 = 5 $ достигается в:
- Минимум в точке $ (4, 1) $ с значением $ z = e^{-5} $.

- Максимум в точке $ (-4, -1) $ с значением $ z = e^{3} $.


### Задание 4

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Пример данных
df = pd.read_csv('exam.csv')

# Преобразование категориальных переменных в числовые
df['Age_Group'] = df['Age_Group'].astype('category').cat.codes
df['Test'] = df['Test'].astype('category').cat.codes
df['Status'] = df['Status'].astype('category').cat.codes

# Разделение данных на признаки и целевую переменную
X = df[['Age_Group', 'Test']]
y = df['Status']

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучение модели Наивного Байеса
model = GaussianNB()
model.fit(X_train, y_train)

# Прогнозирование на тестовой выборке
y_pred = model.predict(X_test)

# Оценка качества модели
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

### Задание 5

```python
import numpy as np

def f(x):
    return x**4 + 3*x**3 - 12*x**2 + 7*x - 2

# Первая производная
def f_prime(x):
    return 4*x**3 + 9*x**2 - 24*x + 7

# Вторая производная
def f_double_prime(x):
    return 12*x**2 + 18*x - 24

# Метод Ньютона
def newton_method(x0, tolerance):
    x_n = x0
    while True:
        f_prime_val = f_prime(x_n)
        f_double_prime_val = f_double_prime(x_n)
        
        # Проверка на ноль для второй производной
        if f_double_prime_val == 0:
            print("Вторая производная равна нулю. Метод Ньютона не может быть применен.")
            return None
        
        # Обновление x
        x_n1 = x_n - f_prime_val / f_double_prime_val
        
        # Проверка на сходимость
        if abs(x_n1 - x_n) < tolerance:
            break
        
        x_n = x_n1
    
    return x_n

# Начальная точка и точность
initial_x = 10
tolerance = 0.0001

# Запуск метода Ньютона
optimal_x = newton_method(initial_x, tolerance)
print(f"Оптимальное значение x: {optimal_x}")
print(f"Значение функции в оптимальной точке: {f(optimal_x)}")
```

```python
# Данные вероятностей
probab = [0.1, 0.3, 0.1, 0.25, 0.05, 0.05, 0.1]

# Вероятность получения 7 спам-писем
sum_probab = sum(probab)
prob_7 = 1 - sum_probab  # Вероятность для 7 писем
probab.append(prob_7)  # Добавим вероятность 7 писем в список

print(f"Вероятность получения 7 спам-писем: {prob_7:.2f}")

# Математическое ожидание
expected_value = 0
for i in range(len(probab)):
    expected_value += i * probab[i]

print(f"Математическое ожидание: {expected_value:.2f}")

# Найдем дисперсию, но сначала найдем E(X^2)
expected_value_squared = 0
for i in range(len(probab)):
    expected_value_squared += (i ** 2) * probab[i]

# Теперь найдем дисперсию
variance = expected_value_squared - (expected_value ** 2)

print(f"Дисперсия: {variance:.2f}")
```

### Задание 7

```python
# Параметр λ
lambda_value = 2

# Функция для вычисления e^(-λ) без использования готовых функций
def exp_neg_lambda(lambda_value):
    e_neg_lambda = 1.0  # Начальное значение для e^(-λ)
    term = 1.0          # Текущий член ряда
    n = 1               # Счетчик для членов ряда
    while n < 100:      # Суммируем первые 100 членов ряда
        term *= -lambda_value / n  # Следующий член ряда
        e_neg_lambda += term
        n += 1
    return e_neg_lambda

# Функция для вычисления k! без использования готовых функций
def factorial(k):
    result = 1
    for i in range(1, k + 1):
        result *= i
    return result

# Функция для вычисления вероятности по распределению Пуассона
def poisson_probability(k, lambda_value):
    e_neg_lambda = exp_neg_lambda(lambda_value)
    return (e_neg_lambda * (lambda_value ** k)) / factorial(k)

# Вычисляем вероятности для k = 2, 3, 4
prob_2 = poisson_probability(2, lambda_value)
prob_3 = poisson_probability(3, lambda_value)
prob_4 = poisson_probability(4, lambda_value)

# Суммируем вероятности
total_probability = prob_2 + prob_3 + prob_4

# Округляем до тысячных
total_probability_rounded = int(total_probability * 1000 + 0.5) / 1000.0

# Выводим результат
print(f"Вероятность, что за час поступит не более 4 и не менее 2 звонков: {total_probability_rounded:.3f}")
```
